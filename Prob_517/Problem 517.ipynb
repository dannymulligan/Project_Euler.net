Problem 517 - A real recursion
====

For every real number $a>1$ is given the sequence $g_a$ by:

$g_a(x)=1$ for $x<a$

$g_a(x)=g_a(x−1)+g_a(x−a)$ for $x>=a$

$G(n)=g_{\sqrt{n}}(n)$

$G(90)=7564511$.

Find $\sum{G(p)}$ for $p$ prime and $10000000<p<10010000$

Give your answer modulo $1000000007$.



```python
import time
start_time = time.clock()
print("time taken {:.1f} seconds".format(time.clock() - start_time))
```

```python
def g(a, x):
    if x < a:
        return 1
    else:
        return g(a, x-1) + g(a, x-a)
```

```python
def G(n):
    return g(n**.5, n)
```

```python
%time G(90)
```
